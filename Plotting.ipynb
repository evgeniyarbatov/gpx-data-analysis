# Make plots

```python
import glob
import os
import gpxpy
import pandas as pd
import plotly.express as px

DATA_DIR = "/Users/arbatov/gitRepo/gpx-data/data/strava"

gpx_files = glob.glob(os.path.join(DATA_DIR,"*.gpx"))

file_contents = []
for gpx_file in gpx_files:
    with open(gpx_file) as file:
        gpx = gpxpy.parse(file)
        
        gpx_points = gpx.tracks[0].segments[0].points
        for point in gpx_points:
            file_contents.append([
                os.path.basename(gpx_file),
                point.time,
                point.longitude, 
                point.latitude,   
                point.elevation
            ])

df = pd.DataFrame(
    file_contents, 
    columns=[
        'filename',
        'time',
        'lon',
        'lat',
        'elev'
    ])

run_data_df.head()
```

Let's make a basic plot

```python
fig_1 = px.scatter(df, x='lon', y='lat', template='plotly_white')
fig_1.show()
```

Elevation plot

```python
fig_2 = px.line(df, x='time', y='elev', template='plotly_white')
fig_2.show()
```

3D plot with elevation

```python
fig_3 = px.scatter_3d(df, x='lon', y='lat', z='elev', color='elev', template='plotly_white')
fig_3.update_traces(marker=dict(size=2), selector=dict(mode='markers'))
fig_3.show()
```

With the map

```python
fig_4 = px.line_mapbox(df, lat='lat', lon='lon', hover_name='time', mapbox_style="open-street-map", zoom=11)
fig_4.update_layout(margin={"r":0,"t":0,"l":0,"b":0})
fig_4.show()
```
